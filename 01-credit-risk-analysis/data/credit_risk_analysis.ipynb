```python
import pandas as pd

df = pd.read_csv("../data/credit_transactions.csv")
df.head()
```

```python
df.info()
df.isnull().sum()
```

```python
df["outstanding_debt"] = df["total_amount"] - df["paid_amount"]

customer_summary = df.groupby("customer_id").agg({
    "total_amount": "sum",
    "paid_amount": "sum",
    "outstanding_debt": "sum",
    "days_past_due": "mean"
}).reset_index()

customer_summary
```
